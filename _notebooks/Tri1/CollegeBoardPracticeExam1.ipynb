## Searched Terms (during the Quiz)

1. **Internet Engineering Task Force (IETF):**
   - *Question*: What is the Internet Engineering Task Force (IETF) and what does it do?
   - *Answer*: The IETF is a community of network designers, operators, vendors, and researchers who work on the evolution and operation of the Internet. It develops and promotes voluntary Internet standards and protocols, including the Request for Comments (RFC) series, which defines many core Internet protocols.

   - *Additional Notes*: The IETF plays a crucial role in shaping the future of the internet by creating and maintaining open standards. These standards are vital for ensuring interoperability and innovation across the global network. The IETF's work is organized into working groups, each focusing on specific internet-related topics, such as routing, security, and communication protocols.

2. **Incremental Process of Program Development:**
   - *Question*: What is the incremental process of program development?
   - *Answer*: Incremental development is an iterative approach to building software. It involves dividing a complex project into smaller, manageable pieces, or increments, and working on each increment one at a time. Developers add features or functionality incrementally, improving the software gradually with each iteration.

   - *Additional Notes*: Incremental development is often used in agile software development methodologies. It allows for early delivery of functional software and provides the flexibility to adapt to changing requirements during the development process. Each increment typically adds new features or refines existing ones, reducing the risk of project failure.

3. **IP Addresses and Internet Devices:**
   - *Question*: Are IP addresses given to every device connected to the Internet?
   - *Answer*: Yes, each device connected to the Internet is typically assigned an IP (Internet Protocol) address. These addresses serve as unique identifiers for devices on the Internet, allowing them to send and receive data.

   - *Additional Notes*: IP addresses can be either IPv4 (32-bit) or IPv6 (128-bit). They play a vital role in routing data across the Internet. Network Address Translation (NAT) is commonly used to manage the allocation of IP addresses in private networks, allowing multiple devices to share a single public IP address.

4. **Overflow Error:**
   - *Question*: What is an overflow error?
   - *Answer*: An overflow error occurs when a computer program or system tries to store a value in a data type that cannot hold it, resulting in unexpected behavior or data corruption.

   - *Additional Notes*: Overflow errors are especially common in programming when working with finite-sized data types like integers. They can lead to security vulnerabilities and software crashes. To mitigate overflow errors, developers use techniques such as bounds checking and data validation.

5. **Undecidable Problem:**
   - *Question*: What is an undecidable problem?
   - *Answer*: An undecidable problem is a computational problem for which there is no algorithm that can always provide a correct answer in a finite amount of time.

   - *Additional Notes*: Undecidability is a fundamental concept in theoretical computer science, demonstrated by problems like the Halting Problem. These problems have implications for issues such as program correctness and the limitations of automated decision-making.

6. **Public Key Cryptography:**
   - *Question*: What is public key cryptography?
   - *Answer*: Public key cryptography uses a pair of keys: a public key and a private key. The public key is used for encryption, while the private key is used for decryption. It enables secure communication and digital signatures.

   - *Additional Notes*: Public key cryptography, also known as asymmetric cryptography, provides a means for secure communication over untrusted networks like the internet. It's the foundation for many security protocols, including HTTPS (for secure web browsing) and SSH (for secure remote access).

7. **Symmetric Encryption:**
   - *Question*: What is symmetric encryption?
   - *Answer*: Symmetric encryption uses the same key for both encryption and decryption. It's faster than asymmetric encryption but requires a secure method for key distribution.

   - *Additional Notes*: Symmetric encryption is efficient for bulk data encryption, making it suitable for tasks like securing files and data at rest. Key management is a critical aspect, and symmetric keys must be securely shared between parties that need to communicate.

8. **Generalization of Procedures:**
   - *Question*: What is the generalization of procedures?
   - *Answer*: This refers to the idea of writing reusable, general procedures or functions that can be applied to various situations or inputs.

   - *Additional Notes*: Generalization is a key principle in software development. It promotes code reusability and maintainability. By creating generalized procedures, developers can solve a wide range of similar problems with a single function, reducing code duplication and errors.

9. **Internet Crowdsourcing:**
   - *Question*: What is internet crowdsourcing?
   - *Answer*: Internet crowdsourcing is a method of obtaining input, ideas, services, or content by soliciting contributions from a large group of people, typically via the internet.

   - *Additional Notes*: Internet crowdsourcing leverages the collective intelligence of online communities to solve problems, generate content, or gather information. Platforms like Wikipedia and Kickstarter are examples of how crowdsourcing is used for collaborative content creation and funding, respectively.

10. **Abstraction:**
    - *Question*: What is the definition of abstraction?
    - *Answer*: Abstraction is a fundamental concept in computer science. It involves simplifying complex systems by focusing on the essential features while ignoring unnecessary details. It allows for easier problem-solving and software design.

   - *Additional Notes*: Abstraction is a cornerstone of software development, allowing programmers to work at higher levels of understanding and hide the complexity of underlying systems. Abstraction can be seen in various forms, such as data abstraction, procedural abstraction, and object-oriented programming.

11. **Cloud Computing:**
    - *Question*: What is cloud computing?
    - *Answer*: Cloud computing is the delivery of on-demand computing services over the internet, such as storage, processing, and software. It allows organizations to access and use resources without the need to invest in and maintain their own infrastructure.

   - *Additional Notes*: Cloud computing has transformed the way businesses and individuals access and manage their computing resources. It offers scalability, cost-efficiency, and flexibility. Cloud services include Infrastructure as a Service (IaaS), Platform as a Service (PaaS), and Software as a Service (SaaS). Popular cloud providers include Amazon Web Services (AWS), Microsoft Azure, and Google Cloud Platform.
