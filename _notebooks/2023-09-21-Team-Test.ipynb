# Program with Iterateration
Create a dictionary of days when school is off with their day numbers, like out of 365

```python
days_off = {
    "New Year's Day": 1,
    "Martin Luther King Jr. Day": 21,
    "President's Day": 52,
    "Spring Break": 100,
    "Memorial Day": 145,
    # Add more days as needed
}
```

Using a for loop to iterate through the dictionary and print as day/month dates

```python
print("Using a for loop to print as day/month dates:")
for day, day_number in days_off.items():
    # Calculate the month and day based on the day_number
    if day_number <= 31:
        month = 1
        day_in_month = day_number
    elif day_number <= 59:
        month = 2
        day_in_month = day_number - 31
    elif day_number <= 90:
        month = 3
        day_in_month = day_number - 59
    # Add more months as needed

    # Print in day/month format
    print(f"{day_in_month}/{month}: {day}")
```

```python
print("\nUsing a while loop to print as month/day dates:")
keys = list(days_off.keys())
index = 0
while index < len(keys):
    day = keys[index]
    day_number = days_off[day]

    # Calculate the month and day based on the day number
    if day_number <= 31:
        month = 1
        day_in_month = day_number
    elif day_number <= 59:
        month = 2
        day_in_month = day_number - 31
    elif day_number <= 90:
        month = 3
        day_in_month = day_number - 59
    # Add more months as needed

    # Print in month/day format
    print(f"{month}/{day_in_month}: {day}")
    index += 1
```

# Code with List

```python
import random


# List of possible answers
answers = ["apple", "banana", "cherry", "date", "elderberry", "fig", "grape", "honeydew", "kiwi", "lemon"]


# Select a random answer from the list
random_answer = random.choice(answers)


# Convert the answer to a list of characters for checking guesses
answer_list = list(random_answer)


# Create a list to store guessed letters
guessed_letters = []


# Number of allowed incorrect guesses
max_attempts = 6
incorrect_attempts = 0


# Display the initial message
print("Welcome to the Guess the Fruit Game!")
print("Can you guess the name of the fruit?")


# Main game loop
while True:
    # Display the current state of the word with underscores for unguessed letters
    display_word = ""
    for letter in random_answer:
        if letter in guessed_letters:
            display_word += letter
        else:
            display_word += "_"
    print(display_word)


    # Prompt the player for a guess
    guess = input("Guess a letter: ").lower()


    # Check if the guess is a single letter
    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a single letter.")
        continue


    # Check if the guess has already been guessed
    if guess in guessed_letters:
        print("You've already guessed that letter.")
        continue


    # Add the guess to the list of guessed letters
    guessed_letters.append(guess)


    # Check if the guess is in the answer
    if guess in random_answer:
        print("Good guess!")
    else:
        print("Incorrect guess.")
        incorrect_attempts += 1


    # Check if the player has won or lost
    if set(guessed_letters) == set(answer_list):
        print(f"Congratulations! You guessed the word: {random_answer}")
        break
    elif incorrect_attempts >= max_attempts:
        print("Sorry, you've run out of attempts. The correct word was:", random_answer)
        break
```

# Code with Selection/Condition


```python
import random


# Generate a random number between 1 and 100
secret_number = random.randint(1, 100)


# Initialize the number of attempts
attempts = 0


while True:
    try:
        # Ask the user to guess the number
        guess = int(input("Guess the secret number (between 1 and 100): "))
       
        # Increment the attempts
        attempts += 1


        # Check if the guess is correct
        if guess == secret_number:
            print(f"Congratulations! You guessed the secret number {secret_number} in {attempts} attempts.")
            break
        elif guess < secret_number:
            print("Try a higher number.")
        else:
            print("Try a lower number.")


    except ValueError:
        print("Please enter a valid number.")
```

# Code with Purpose

```python


#calendar making


import calendar


# Input month and year
year = 2008
month = 1 # September


# Create a dictionary to map month numbers to month names
months = {
1: "January", 2: "February", 3: "March", 4: "April",
5: "May", 6: "June", 7: "July", 8: "August",
9: "September", 10: "October", 11: "November", 12: "December"
}


# Create a calendar for the specified month and year
cal = calendar.monthcalendar(year, month)


# Create a list of day names
day_names = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]


# Print the calendar header
print(f"Calendar for {months[month]} {year}")
print(" ".join(day_names))


# Iterate through the weeks in the calendar and print the days
for week in cal:
    week_str = ""
    for day in week:
        if day == 0:
            week_str += " " # Print empty space for days outside the month
        else:
            week_str += f"{day:2d} " # Print day with leading space if needed
    print(week_str)
```
